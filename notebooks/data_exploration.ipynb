# Co-appearance Matrix for Multilabel Spans
This part demonstrates how to compute and visualize the co-appearance (co-occurrence) matrix for multilabel spans in the dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations
import sys
sys.path.append('../scripts')  # adjust path as needed
```

## Load Label Data
Adjust the file path below to point to a valid label file (e.g., for English: 'data/raw/en/train-labels-subtask-3-spans.txt').

```python
label_file = '../data/raw/en/train-labels-subtask-3-spans.txt'  # Change as needed
df = pd.read_csv(label_file, sep='\t', header=None, names=['doc_id', 'label', 'start', 'end'])
df.head()
```

## Compute Co-appearance Matrix

```python
from data_exploration.multilabel_patterns import compute_cooccurrence_matrix, plot_cooccurrence_matrix, compute_exact_span_cooccurrence_matrix

co_matrix = compute_cooccurrence_matrix(df)
exact_co_matrix = compute_exact_span_cooccurrence_matrix(df)
# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Plot the co-occurrence matrices on the respective axes
plot_cooccurrence_matrix(co_matrix, title='Co-occurrence Matrix', ax=ax1)
plot_cooccurrence_matrix(exact_co_matrix, title='Exact Span Co-occurrence Matrix', ax=ax2)

# Adjust layout to prevent overlap
plt.tight_layout()
```
